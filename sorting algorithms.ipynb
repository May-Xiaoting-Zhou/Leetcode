# Selection Sort
### The algorithm repeatedly selects the smallest (or largest) element from the unsorted portion of the list and swaps it with the first element of the unsorted part. This process is repeated for the remaining unsorted portion until the entire list is sorted. 

```python
idx_min = 0
idx_sorted = 0
nums = [3, 5, 6, 9, 10, 34, 0.5, 1, 2, 78, 67, 55, 33, 999]
while idx_sorted < len(nums):
    min_n = nums[idx_sorted]
    idx_min = idx_sorted
    for i in range(idx_sorted, len(nums)-1):
        if nums[i+1] < min_n:
            min_n = nums[i+1]
            idx_min = i + 1
    nums[idx_min], nums[idx_sorted] = nums[idx_sorted], nums[idx_min]
    idx_sorted += 1
nums_lst.sort()
assert nums == nums_lst, "nums should be sorted"
```

# Bubble Sort
### In Bubble Sort algorithm, 
- traverse from left and compare adjacent elements and the higher one is placed at right side. 
- In this way, the largest element is moved to the rightmost end at first. 
- This process is then continued to find the second largest and place it and so on until the data is sorted.


```python
nums = [3, 5, 6, 9, 10, 34, 0.5, 1, 2, 78, 67, 55, 33, 999]

idx_sorted = len(nums)-1

while idx_sorted > 0:
    swapped = False
    for i in range(idx_sorted):
        if nums[i] > nums[i+1]:
            nums[i], nums[i+1] = nums[i+1], nums[i]
            swapped = True
    if swapped == False:
        break
    idx_sorted -= 1
nums_lst.sort()
assert nums == nums_lst, "nums should be sorted"
```

```python
# Optimized Python program for implementation of Bubble Sort
def bubbleSort(arr):
    n = len(arr)
    
    # Traverse through all array elements
    for i in range(n):
        swapped = False

        # Last i elements are already in place
        for j in range(0, n-i-1):

            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if (swapped == False):
            break


# Driver code to test above
if __name__ == "__main__":
    arr = [64, 34, 25, 12, 22, 11, 90]

    bubbleSort(arr)

    print("Sorted array:")
    for i in range(len(arr)):
        print("%d" % arr[i], end=" ")

# This code is modified by Suraj krushna Yadav
```

# Insertion Sort
### Insertion sort is like sorting playing cards in your hands. You split the cards into two groups: the sorted cards and the unsorted cards. Then, you pick a card from the unsorted group and put it in the right place in the sorted group.


```python
nums = [3, 5, 6, 9, 10, 34, 0.5, 1, 2, 78, 67, 55, 33, 999]

idx_sorted = 1

while idx_sorted < len(nums):
    val_insert = nums[idx_sorted]
    
    for i in range(0, idx_sorted):

        if val_insert < nums[i]:
            nums[i+1:idx_sorted+1] = nums[i:idx_sorted]
            nums[i] = val_insert
            break
    idx_sorted += 1

nums_lst.sort()
assert nums == nums_lst, "nums should be sorted"
```

```python
# Python program for implementation of Insertion Sort

# Function to do insertion sort
def insertionSort(arr):

    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):

        key = arr[i]

        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >= 0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key


# Driver code to test above
arr = [12, 11, 13, 5, 6]
insertionSort(arr)
for i in range(len(arr)):
    print ("% d" % arr[i])

# This code is contributed by Mohit Kumra
```

# Merge Sort
### Merge sort is a sorting algorithm that follows the divide-and-conquer approach. It works by recursively dividing the input array into smaller subarrays and sorting those subarrays then merging them back together to obtain the sorted array.


```python
def merge_sort(nums1, nums2):
    for i in range(len(nums1)):
        j = 0
        if nums1[i] >= nums2[j]:
            nums1[i], nums2[j] = nums2[j], nums1[i]
            for k in range(j+1, len(nums2), 1):
                if nums2[j] > nums2[k]:
                    nums2[j], nums2[k] = nums2[k], nums2[j] 
                    j = k 
                else:
                    break
    return nums1 + nums2

def divide_sort(nums):
    if len(nums) == 1:
        return nums
    elif len(nums) == 2:
        min_val = min(nums)
        max_val = max(nums)
        nums[0], nums[1] = min_val, max_val
        return nums
    else:
        nums = merge_sort(divide_sort(nums[:len(nums)//2]), divide_sort(nums[len(nums)//2:]))
        return nums
```

```python
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m
    L = [0] * n1
    R = [0] * n2

    for i in range(n1):
        L[i] = arr[l + i]
    for j in range(n2):
        R[j] = arr[m + 1 + j]

    i, j, k = 0, 0, l
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

def mergeSort(arr, l, r):
    if l < r:
        m = l + (r - l) // 2
        mergeSort(arr, l, m)
        mergeSort(arr, m + 1, r)
        merge(arr, l, m, r)

arr = [12, 11, 13, 5, 6, 7]
n = len(arr)

print("Given array is:", end=" ")
for i in range(n):
    print(arr[i], end=" ")

mergeSort(arr, 0, n - 1)

print("\n\nSorted array is:", end=" ")
for i in range(n):
    print(arr[i], end=" ")
```
